### Tentando calcular a FFT ###

```python
import numpy as np
from matplotlib import pyplot as plt
from cmath import polar
```

```python
formas = np.loadtxt('src/formasEditado
                    .txt', delimiter=",")
```

```python
fo1 = np.array(formas[0])
fo2 = np.array(formas[1])
fo3 = np.array(formas[2])
fo4 = np.array(formas[3])
```

```python
print(fo1)
print(fo2)
print(fo3)
print(fo4)
```

```python
fft_1 = np.fft.fft(fo1,n=64,norm="backward")
print(fft_1[:10])
```

```python
fft_1 = np.fft.fft(fo1,n=64,norm="forward")
#fft_1
np.abs(fft_1)
plt.plot(np.abs(fft_1))
```

```python
# FFt com array de reais usando normalização backward
# oresultado na poisiçao 1 é 160 (não faço ideia do significado)
fft_2 = np.fft.rfft(fo1,n=64,norm="backward")
print(fft_2)
```

```python
ondaArtificial = np.fft.irfft(fft_2)
plt.plot(ondaArtificial)
```

```python
plt.plot(fo1)
```

```python
# FFT com array de reais usando normalização forward
# resultado na poisção 1 (60Hz) corresponde ao valor gerado
fft_2 = np.fft.rfft(fo1,n=64,norm="forward")
np.abs(fft_2)
```

```python
fft_2= np.fft.fft(fo2,n=64,norm="forward")

for z in fft_2[:13]:
    p = polar(z)
    print(round(p[0],2))
```

```python
fft_3 = np.fft.fft(fo3,n=64,norm="forward")

for z in fft_3[:13]:
    p = polar(z)
    print(round(p[0],2))
```
